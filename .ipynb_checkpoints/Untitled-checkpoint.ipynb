# Exploratory data analysis on Titanic dataset by rohan

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = "titanic_data_set.csv"
```

```python
df = pd.read_csv(data)
```

```python
df.shape
```

```python
df.head()
```

```python
df.tail()
```

```python
df.describe()
```

```python
df.describe(include=['O'])
```

```python
df.isnull().sum()
```

```python
df.Cabin.value_counts()
```

```python
df = df.drop(['Cabin'], axis = 1)
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df.isnull().sum()
```

```python
df.Embarked.value_counts()
```

```python
df['Embarked'] = df['Embarked'].fillna('S')
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df['Familysize'] = df['Parch'] + df['SibSp'] + 1
```

```python
df.head()
```

```python
df['Gender'] = df.Sex.apply(lambda x:1 if x=='male' else 0)
```

# Univariate Analysis (visualisations)

```python
g = sns.countplot(x = 'Survived', data = df)
g.set_xlabel('1 = Survived, 0 = Dead')
g.set_title("Survival count plot")
sns.despine()
```

```python
g1 = sns.countplot(x = 'Sex', data = df)
g1.set_title("Male female proportions on the ship")
```

```python
sns.boxplot(y = 'Age', data = df)
```

```python
p = sns.countplot(x = 'Embarked', data = df)
```

# Bivariate Analysis (visualisations)

```python
sns.jointplot(x = 'Age', y = 'Gender', data = df)
```

```python
sns.barplot(x = 'Pclass', y = 'Survived', data = df)
```

```python
sns.barplot(x = 'Sex', y = 'Survived', data = df)
```
